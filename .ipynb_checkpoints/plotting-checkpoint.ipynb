```python
import numpy as np

#get some round's data
scores = np.genfromtxt("data/round8_observed_scores_scalarized.txt")
```

```python
from matplotlib import pyplot as plt

plt.plot(range(len(scores)), scores)
plt.xlabel("Number of Points Sampled (increases with more rounds)")
plt.ylabel("Scalarized Score")
plt.title("Score by Number of Points for the Final Scalarization")
```

```python
average_scores = np.genfromtxt("average_scores.txt")
```

```python
average_scores2 = average_scores[0:10]
```

```python
average_scores2
```

```python
plt.plot(range(len(average_scores2[:,0])), average_scores2[:,0])
plt.xlabel("Round Number")
plt.ylabel("Average VAE scores")
plt.title("Round Number vs. VAE Score")
plt.show()

plt.plot(range(len(average_scores2[:,1])), average_scores2[:,1])
plt.xlabel("Round Number")
plt.ylabel("Average Pyrosetta score (cost * -1)")
plt.title("Round Number vs. Pyr Score")
plt.show()


plt.plot(range(len(average_scores2[:,2])), average_scores2[:,2])
plt.xlabel("Round Number")
plt.ylabel("Average DCA scores")
plt.title("Round Number vs. DCA Score")
plt.show()

plt.plot(range(len(average_scores2[:,3])), (-1)*average_scores2[:,3])
plt.xlabel("Round Number")
plt.ylabel("Average Hamming scores (cost * -1)")
plt.title("Round Number vs. Hamming Score")

```
