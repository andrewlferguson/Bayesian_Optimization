```python
import numpy as np

#get some round's data
scores = np.genfromtxt("data/round29_observed_scores.txt")
#by column: VAE, Pyrosetta, DCA, Functionality
```

```python
from matplotlib import pyplot as plt

#plot the VAE probability across rounds
plt.plot(range(len(scores)), (-1) * scores[:,0])
plt.xlabel("Number of Points Sampled (increases with more rounds)")
plt.ylabel("VAE Probability")
plt.title("Score by Number of Points for the Final Scalarization")
```

```python
#plot the Pyrosetta stability score across rounds
plt.plot(range(len(scores)), scores[:,1])
plt.xlabel("Number of Points Sampled (increases with more rounds)")
plt.ylabel("Pyrosetta Energy")
plt.title("Score by Number of Points for the Final Scalarization")
```
